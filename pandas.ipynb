# series

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
s =pd.Series(np.random.randn(5),index = ['a','b','c','d','e'])
print(s)
```

```python
s =pd.Series(np.random.randn(5))
s
```

```python
s =pd.Series(np.random.randn(5))
print(s)
```

```python
# created dictionary
d1 = {'b':1,'a':0,'c':2}
s1 = pd.Series(d1)
print(s1)
```

```python
d1 = {'b':1,'a':0,'c':2}
i1 = ['p','q','r']
s2 = pd.Series(d1,index =i1)
print(s2)
```

```python
d1 = {'b':1,'a':0,'c':2}
i1 = ['a','b','c']
s2 = pd.Series(d1,index =i1)
print(s2)
```

```python
#scaler value
i3 = ['a', 'b', 'c', 'd', 'e']
s3 = pd.Series(3., index = i3)
print(s3)
```
